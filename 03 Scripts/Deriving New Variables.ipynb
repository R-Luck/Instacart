## Question 2

```python
# Check how many orders occur on what day of the week
df_merged_large['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create loop to assign values to 2 busiest days and 2 slowest days

result = []
for value in df_merged_large['orders_day_of_week']:
    if value == 0:
        result.append("Busiest days")
    elif value == 1:
        result.append("Busiest days")
    elif value == 3:
        result.append("Slowest days")
    elif value == 4:
        result.append("Slowest days")
    else: 
        result.append("Regularly busy")
```

```python
# Show result
result
```

```python
# Create new column in df with result list
df_merged_large['busiest_days'] = result
```

```python
# Check frequency of new column
df_merged_large['busiest_days'].value_counts(dropna = False)
```

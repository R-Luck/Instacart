### 3.3 Explore Data Types of Columns

```python
# Print data types of columns
customers.dtypes
```

```python
# Change data type of user id column from integer to string
customers['user_id'] = customers['user_id'].astype(str)
```

```python
# Print data type of user id column
customers['user_id'].dtype
```

```python
# Check for mixed data type columns
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# Change data type of First Name column to string
customers['First Name'] = customers['First Name'].astype(str)
```

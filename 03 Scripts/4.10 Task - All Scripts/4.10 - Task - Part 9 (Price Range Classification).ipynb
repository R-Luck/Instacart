# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# 2. Import Data

```python
# Creating path string to folder
path = r'C:\Users\Admin\Desktop\Instacart Basket Analysis'
```

```python
# Importing df from csv
active_customer = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'active_customer_full_profiles_dep.csv')) 
```

```python
# Check shape of df
active_customer.shape
```

# 3. Derive New Variable - Price Range

```python
# Categorise products as low, mid or high range using price parameters and loc() function
active_customer.loc[active_customer['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
active_customer.loc[(active_customer['prices'] <= 15) & (active_customer['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
active_customer.loc[active_customer['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Check frequency of new column
active_customer['price_range_loc'].value_counts(dropna = False)
```

```python
# create bar chart
price_range_bar = active_customer['price_range_loc'].value_counts().plot.bar()
plt.title('Distribution of Products by Price Range')
```

```python
# Export chart as image file
price_range_bar.figure.savefig(os.path.join(path, '04 Analysis','Visualisations','price_range_bar.png'),bbox_inches='tight')
```

# 4. Price Range and Customer Profile Analysis

#### 4.1 Income Profile

```python
# Create crosstab between 'price_range_loc' and 'income_profile'
crosstab_income = pd.crosstab(active_customer['price_range_loc'],active_customer['income_profile'],dropna = False)
```

```python
# Copy crosstab to cipboard
crosstab_income.to_clipboard()
```

```python
# Create stacked bar chart
stacked_income_price_range = crosstab_income.plot.bar(stacked = True)
plt.legend(title = 'Income Level')
plt.title('Price Range by Income Level')
```

```python
# Create grouped bar chart
grouped_income_price_range = crosstab_income.plot.bar(rot = 0)
plt.legend(title = 'Income Level')
plt.title('Price Range by Income Level')
```

```python
# Export chart as image file
grouped_income_price_range.figure.savefig(os.path.join(path, '04 Analysis','Visualisations','grouped_income_price_range.png'),bbox_inches='tight')
```

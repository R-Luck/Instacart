### 3.1 Create Subsets

```python
# Create two subsets from ords_prods_merge
ords_prods_merge_1 = ords_prods_merge[:16000000]
```

```python
ords_prods_merge_2 = ords_prods_merge[16000000:]
```

### 3.2 Merge Subsets with customers df

```python
# Merge customers and ords_prods_merge_1 dataframes to create df_combined_1
df_combined_1 = ords_prods_merge_1.merge(customers, on = 'user_id', indicator = True)
```

```python
# Merge customers and ords_prods_merge_2 dataframes to create df_combined_2
df_combined_2 = ords_prods_merge_2.merge(customers, on = 'user_id', indicator = True)
```

```python
df_combined_2.shape
```

### 3.3 Append subsets in to new df

```python
# Append df_combined_2 on to df_combined_1 to make new df
df_combined = df_combined_1.append(df_combined_2)
```

### Create Diet/Purchasing Profile - based on department id's

```python
# Create loop to create new object 'result' based on department id's

result = []

for value in test_customer['department_id']:
    if value == 12:
        result.append('Meat Eater')
    elif value == 8:
        result.append('Pet Owner')
    elif value == 5:
        result.append('Alcohol Drinker')
    else:
        result.append('No Profile')
```

```python
# Assign result to new column
test_customer['Department Profile'] = result
```

```python
# Frequency check new column
test_customer['Department Profile'].value_counts(dropna = False)
```

```python
# Create subset of 'Pet Owner'
test_customer_pet_owner = test_customer.loc[test_customer['Department Profile'] == 'Pet Owner']
```

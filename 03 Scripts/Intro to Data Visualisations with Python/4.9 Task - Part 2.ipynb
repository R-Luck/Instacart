### Question 3 - Histogram of 'order_hour_of_day' column

```python
# Create histogram of 'hour order placed'
ords_prods_merge['hour_order_placed'].plot.hist(bins = 24)
```

The above histogram shows that the majority of orders are placed between 9am and 4pm. Orders steadily slow down from 4pm until 1am, with very few orders being placed in the early hours of the morning. We can see orders start to increase again at 5am, when they begin to rise steadily towards the beginning of the peak order window of 9am - 4pm.

```python
# Save histogram as object hist
hist = ords_prods_merge['hour_order_placed'].plot.hist(bins = 24)
```

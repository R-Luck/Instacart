## Question 2

```python
# Find aggregated mean of 'order_number' column when grouped by 'department_id'
ords_prods_merge.groupby('department_id').agg({'order_number':['mean']})
```

## Question 3

The results differ from those produced by the subset as we have a mean for every department. Our subset only produced mean values for 8 out of 21 of the departments. By analysing our entire dataframe we are able to get a truer understanding of which departments perform better than others, our data is not skewed by the subset we created. 

Our subset data only showed a mean value of around 11 for department 17 (household goods). By looking at our whole data set we can see that the mean number of orders for this department is actually closer to 16, in other words this department peforms better than our subset would suggest.

## Question 4

```python
# Create a column 'max order' to show maximum number of orders by each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check df
ords_prods_merge.head()
```

```python
# Use loc() function to create customer loyalty flag based on maximum orders per user
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check frequency of loyalty_flag column
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check head of relevant columns
ords_prods_merge[['user_id','loyalty_flag','order_number']].head(30)
```

## Question 5.

```python
# Find aggregated mean of 'prices' column when grouped by 'loyalty_flag'
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean']})
```

## Question 6.

```python
# Create a column 'mean product price' to show mean price of product purchased by each user
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check head of relevant columns
ords_prods_merge[['user_id','mean_product_price','prices']].head(30)
```

```python
# Use loc() function to create customer spending flag based on mean prices of product per user
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Check output
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## Question 7.

```python
# Create a column 'median days since last order' to show median time since last order by each user
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check head of relevant columns
ords_prods_merge[['user_id','median_days_since_prior_order','days_since_prior_order']].head(30)
```

```python
# Use loc() function to create customer frequency flag based on median days since last order 
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequency of new column
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

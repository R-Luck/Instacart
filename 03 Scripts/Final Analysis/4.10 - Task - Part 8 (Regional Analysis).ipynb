# 3. Regional Analysis by Customer Profile

#### 3.1 Income Profile

```python
# Create crosstab between 'Region' and 'income_profile'
crosstab_income = pd.crosstab(active_customer['Region'],active_customer['income_profile'],dropna = False)
```

```python
# Copy crosstab to cipboard
crosstab_income.to_clipboard()
```

```python
# Create stacked bar chart
stacked_income_region = crosstab_income.plot.bar(stacked = True)
plt.legend(title = 'Income Level')
plt.title('Customer Income Level by Region')
```

```python
# Export chart as image file
stacked_income_region.figure.savefig(os.path.join(path, '04 Analysis','Visualisations','stacked_income_region.png'),bbox_inches='tight')
```

```python
# Create grouped bar chart
grouped_income_region = crosstab_income.plot.bar(rot = 0)
plt.legend(title = 'Income Level')
plt.title('Customer Income Level by Region')
```

```python
# Export chart as image file
grouped_income_region.figure.savefig(os.path.join(path, '04 Analysis','Visualisations','grouped_income_region.png'),bbox_inches='tight')
```
